```python
from sklearn.linear_model import LinearRegression
import requests
import pandas as pd
import numpy as np
```

Fetching Historical data from binance

```python
# fetching historical data for BTC-USD for 1m timeframe for the last 1000 
url = "https://api.binance.com/api/v3/klines?symbol=BTCUSDT&interval=1m&limit=1000"
response = requests.get(url)
```

```python
# printing the response to check if we got the data
print(response.json())
```

```python
# conver the data into json
data = response.json()
```

```python
print( data)
```

Extracting close prices from the data

```python
# extracting the close prices from the data
close_prices = []

for row in data:
    close_prices.append(float(row[4]))
```

```python
print(close_prices[0])
print(len(close_prices))
print(type(close_prices[0]))
```

```python
# printing first 5 and last 5 values of close_prices
print(close_prices[:5])
print(close_prices[-5:])
```

```python
# checking NaN values in close_prices
print(np.isnan(close_prices).any())
```

Normalizing each value between 0 and 1 so that model learns easily and do not overfit

```python
# normalizing the close prices
close_prices = np.array(close_prices)
print(close_prices.max())
print(close_prices.min())

normalized_closed_prices = []
for value in close_prices:
    normalized_value =  (value-close_prices.min())/(close_prices.max()-close_prices.min())
    normalized_closed_prices.append(normalized_value)

```

```python
normalized_closed_prices = np.array(normalized_closed_prices)

print( "Minimum value in normalized closed prices: ", normalized_closed_prices.min())
print( "Maximum value in normalized closed prices: ", normalized_closed_prices.max())

# printing first 5 and last 5 values of normalized_closed_prices
print(normalized_closed_prices[:5])
print(normalized_closed_prices[-5:])
```

Using Sliding Window technique to create X's and Y's for model training

```python
# sliding window technique to create features and target variable
x_feature=[]
y_target=[]

# now applying the sliding window technique to create features and target variable
window_size = 5
for i in range(len(normalized_closed_prices)-window_size):
    x_feature.append(normalized_closed_prices[i:i+window_size])
    y_target.append(normalized_closed_prices[i+window_size])

print(f' features are : {x_feature}')
print(f' target values are : {y_target}')

```

```python
print(f' length of features is : {len(x_feature)}')
print(f' length of target values is : {len(y_target)}')
print("first sample x ",x_feature[0])
print("first sample y ",y_target[0])
# making them nump
x_feature = np.array(x_feature)
y_target = np.array(y_target)
print(type(x_feature))
print(type(y_target))
```

```python
# train the model
model = LinearRegression()
model.fit(x_feature, y_target)
```
